```python
from ultralytics import YOLO
import cv2

model = YOLO('/home/prem/Downloads/yolov8x-pose.pt')
cap = cv2.VideoCapture("/home/prem/Documents/Work/Football/videos/2e57b9_0.mp4")


fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter("output.mp4", fourcc, cap.get(cv2.CAP_PROP_FPS),
                      (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)
    annotated = results[0].plot()
    out.write(annotated)

cap.release()
out.release()
```
